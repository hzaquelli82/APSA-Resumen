```python
import MySQLdb

db = MySQLdb.connect(host="192.168.0.12",    # tu host, usualmente localhost
                     user="operario",         # tu usuario
                     passwd="",  # tu password
                     db="dbp8100")        # el nombre de la base de datos

# Debes crear un objeto Cursor. Te permitirá
# ejecutar todos los queries que necesitas
cur = db.cursor()

# Usa todas las sentencias SQL que quieras
cur.execute("SELECT productox.Nombre AS Producto FROM productox")
```

```python
#Extraer la tabla de productos
cur.execute("SELECT productox.Nombre AS Producto FROM productox ORDER BY productox.Nombre")
productos = cur.fetchall()
cur.close()

#Crear una lista con los nombres de los productos
lista_productos = [prod[0] for prod in productos]
print(len(lista_productos))

#Eliminar nombres de productos duplicados
lista_productos = list(set(lista_productos))
print(len(lista_productos))

```

```python
import pandas as pd

# Cargar el archivo de Excel
df = pd.read_excel("codigos.xlsx")

# Mostrar las primeras filas del DataFrame
print(df.head())
```

```python
productos_jvp = df['Producto'].tolist()

valores_faltantes = [producto for producto in lista_productos if producto not in productos_jvp]

valores_faltantes
```

```python
for producto in valores_faltantes:
    nueva_fila = pd.DataFrame({"Producto": [producto], "Cod": [None]})
    df = pd.concat([df, nueva_fila], ignore_index=True)

# Ordenamos el DataFrame por la columna 'Producto'
df = df.sort_values(by='Producto')

print(df)
```
